```python
import numpy as np
import pandas as pd

# Leer todos los archivos
dfs = []

for year in range(2014,2027):
    df = pd.read_csv(f"../Data/Raw data/{year}_LoL_esports_match_data_from_OraclesElixir.csv",low_memory=False)
    dfs.append(df)

# Unir todos
datos = pd.concat(dfs, ignore_index=True)

# Crear lista global de campeones
campeones = sorted(datos["champion"].dropna().unique())

# Crear diccionario
champion_map = {champ: i for i, champ in enumerate(campeones)}

# Aplicarlo
datos["champion_id"] = datos["champion"].map(champion_map)

# Convertir también a categoría
datos["champion"] = pd.Categorical(
    datos["champion"],
    categories=campeones
)

# Guardar el diccionario
mapping = pd.DataFrame({
    "champion": campeones,
    "champion_id": range(len(campeones))
})

mapping.to_csv("../Data/Aux data/champion_mapping.csv", index=False)

print(mapping.head())
```
