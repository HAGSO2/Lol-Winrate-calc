```python
import numpy as np
import pandas as pd
```

```python
year = "2021"
df = pd.read_csv(f"../Data/Raw Data/{year}_LoL_esports_match_data_from_OraclesElixir.csv",low_memory=True)
```

```python
def re_arrange_year(my_df, this_year):
    filas = []
    
    for gameid, partida in my_df.groupby("gameid"):
    
        equipos = partida[partida["position"] == "team"]
        jugadores = partida[partida["position"] != "team"]
    
        for side in ["Blue", "Red"]:
    
            row = {
                "gameid": gameid,
                "side": side
            }
    
            team = equipos[equipos["side"] == side].iloc[0]
    
            posiciones = ["top","jng","mid","bot","sup"]
    
            for i, pos in enumerate(posiciones, start=1):
    
                p = jugadores[
                    (jugadores["side"] == side) &
                    (jugadores["position"] == pos)
                ].iloc[0]
    
                row[f"p{i}_champion"] = p["champion"]
                row[f"p{i}_kills"] = p["kills"]
                row[f"p{i}_deaths"] = p["deaths"]
                row[f"p{i}_pentakill"] = p["pentakills"]
                row[f"p{i}_dpm"] = p["dpm"]
    
            row["dragons"] = team["dragons"]
            row["barons"] = team["barons"]
            row["heralds"] = team["heralds"]
            row["towers"] = team["towers"]
            row["result"] = team["result"]
    
            filas.append(row)
    
    resultado = pd.DataFrame(filas)
    resultado.to_csv(f"../Data/Procesed data/Cleaned data/{this_year}_LoL_esports_match_data_from_OraclesElixir.csv", index=False)
```

```python
years = list(range(2014,2027))

for y in years:
    this_df =  pd.read_csv(f"../Data/Raw Data/{str(y)}_LoL_esports_match_data_from_OraclesElixir.csv",low_memory=True)
    re_arrange_year(this_df,y)
    print(f"year {y}, done!")
```
