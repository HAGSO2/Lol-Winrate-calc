```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
año = "2024"
df = pd.read_csv(f"../Data/Raw Data/{año}_LoL_esports_match_data_from_OraclesElixir.csv",low_memory=False)
```

```python
# Mantener únicamente los jugadores
df = df[df["participantid"].between(1, 10)]

# Eliminar valores nulos
df = df.dropna(subset=["position", "dpm"])

# Orden de los roles
roles = ["top", "jng", "mid", "bot", "sup"]

# Obtener los datos de DPM para cada rol
datos = [
    df[df["position"] == rol]["dpm"]
    for rol in roles
]

# Crear boxplot
plt.figure(figsize=(10, 6))

plt.boxplot(
    datos,
    labels=["Top", "Jungla", "Mid", "ADC", "Support"],
    patch_artist=True,
    showmeans=True
)

plt.title("Daño a campeones por minuto según el rol")
plt.xlabel("Rol")
plt.ylabel("Daño por minuto (DPM)")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
