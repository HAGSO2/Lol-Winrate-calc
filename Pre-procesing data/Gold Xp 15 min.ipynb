```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
año = "2024"
df = pd.read_csv(f"../Data/Raw Data/{año}_LoL_esports_match_data_from_OraclesElixir.csv",low_memory=False)
```

```python
# Mantener únicamente las filas de equipos
df = df[df["participantid"].isin([100, 200])]

# Eliminar valores nulos
df = df.dropna(subset=["golddiffat15", "xpdiffat15", "result"])

# Separar victorias y derrotas
victorias = df[df["result"] == 1]
derrotas = df[df["result"] == 0]

fig, ax = plt.subplots(1, 2, figsize=(12,6))

# Diferencia de oro
ax[0].boxplot(
    [derrotas["golddiffat15"], victorias["golddiffat15"]],
    labels=["Derrota", "Victoria"]
)
ax[0].set_title("Diferencia de oro a los 15 min")
ax[0].set_ylabel("Oro")

# Diferencia de experiencia
ax[1].boxplot(
    [derrotas["xpdiffat15"], victorias["xpdiffat15"]],
    labels=["Derrota", "Victoria"]
)
ax[1].set_title("Diferencia de XP a los 15 min")
ax[1].set_ylabel("Experiencia")

plt.suptitle("Impacto de la ventaja a los 15 minutos")
plt.tight_layout()
plt.show()
```

```python
año = "2024"
df = pd.read_csv(f"../Data/Raw Data/{año}_LoL_esports_match_data_from_OraclesElixir.csv",low_memory=False)
```

```python
df = df[df["participantid"].isin([100, 200])]
df = df.dropna(subset=["golddiffat15", "xpdiffat15", "result"])

victorias = df[df["result"] == 1]
derrotas = df[df["result"] == 0]

fig, ax = plt.subplots(1, 2, figsize=(12,6))

# Oro
ax[0].violinplot(
    [derrotas["golddiffat15"], victorias["golddiffat15"]],
    showmeans=True,
    showmedians=True
)
ax[0].set_xticks([1,2])
ax[0].set_xticklabels(["Derrota", "Victoria"])
ax[0].set_title("Diferencia de oro a los 15 min")
ax[0].set_ylabel("Oro")

# XP
ax[1].violinplot(
    [derrotas["xpdiffat15"], victorias["xpdiffat15"]],
    showmeans=True,
    showmedians=True
)
ax[1].set_xticks([1,2])
ax[1].set_xticklabels(["Derrota", "Victoria"])
ax[1].set_title("Diferencia de XP a los 15 min")
ax[1].set_ylabel("Experiencia")

plt.suptitle("Impacto de la ventaja a los 15 minutos")
plt.tight_layout()
plt.show()
```
